```python
import pandas as pd

# Reloading the CSV with updated method for StringIO
df = pd.read_csv("/Users/jasminehuang/Desktop/datasets to be used/MBTA_Rapid_Transit_Stop_Distances.csv")

# Group by route_pattern_name and collect station IDs in the path order
route_paths = df.groupby('route_pattern_name').apply(
    lambda g: g['from_station_id'].tolist() + [g['to_station_id'].iloc[-1]]
).tolist()


```

```python
import json
with open('station-paths.json', 'w') as f:
    json.dump(route_paths, f, indent=4)
```
