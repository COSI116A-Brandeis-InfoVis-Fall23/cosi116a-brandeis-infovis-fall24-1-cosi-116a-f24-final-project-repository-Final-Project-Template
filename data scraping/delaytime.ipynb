```python
import pandas as pd
import numpy as np
```

```python
data_dir = "TravelTimes_2024"  # Replace with the directory containing your files

# Step 1: Load and Combine All Files
all_data = []
for file in os.listdir(data_dir):
    if file.endswith(".csv"):
        file_path = os.path.join(data_dir, file)
        df = pd.read_csv(file_path)
        all_data.append(df)

# Combine all DataFrames into one
df = pd.concat(all_data, ignore_index=True)

# Ensure travel_time_sec is numeric
df['travel_time_sec'] = pd.to_numeric(df['travel_time_sec'], errors='coerce')
df['from_stop_id'] = df['from_stop_id'].astype(str)
df['to_stop_id'] = df['to_stop_id'].astype(str)
```

```python
unique_stops = sorted(set(df['from_parent_station']).union(set(df['to_parent_station'])))
stop_index = {stop: idx for idx, stop in enumerate(unique_stops)}  # Map stop IDs to indices

# Initialize 2D arrays with NaN (to represent no data)
n = len(unique_stops)
min_travel_times = np.full((n, n), np.nan)
avg_travel_times = np.full((n, n), np.nan)

grouped = df.groupby(['from_parent_station', 'to_parent_station'])['travel_time_sec']

min_times = grouped.min()
avg_times = grouped.mean()

# Fill the 2D arrays
for (from_stop, to_stop), min_time in min_times.items():
    i, j = stop_index[from_stop], stop_index[to_stop]
    min_travel_times[i, j] = min_time

for (from_stop, to_stop), avg_time in avg_times.items():
    i, j = stop_index[from_stop], stop_index[to_stop]
    avg_travel_times[i, j] = avg_time

# Convert NumPy arrays to DataFrames with stop IDs as labels
min_travel_times_df = pd.DataFrame(min_travel_times, index=unique_stops, columns=unique_stops)
avg_travel_times_df = pd.DataFrame(avg_travel_times, index=unique_stops, columns=unique_stops)
```

```python
avg_travel_times_df
```

```python
delay_matrix = avg_travel_times_df.values - min_travel_times_df.values

# Mask NaN values to avoid them in calculations
delay_matrix = np.where(np.isnan(delay_matrix), np.nan, delay_matrix)

# Calculate the average delay for each station (ignoring NaNs)
average_delay_per_station = np.nanmean(delay_matrix, axis=1)

# Create a DataFrame for the results
average_delay_df = pd.DataFrame({
    'stop_id': min_travel_times_df.index,  # Station labels
    'average_delay_sec': average_delay_per_station
})
```

```python
print(average_delay_df)
average_delay_df.to_csv("average_delay_per_station.csv", index=False)
```
